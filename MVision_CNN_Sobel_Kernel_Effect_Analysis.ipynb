```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib.image as mping

import cv2

image = mping.imread('curved_lane.jpg')
image1  = mping.imread('bridge_trees_example.jpg')

f , sub =  plt.subplot(1,2)
sub[1,0].imshow(image)
sub[1,1].imshow(image1)
```

```
# let's convert image to gray scale 
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap = 'gray')

gray1 = cv2.cvtColor(image1, cv2.COLOR_RGB2GRAY)
plt.imshow(gray1, cmap = 'gray')

```

###  Now theoretical process use sobel kernel and apply it to image as filter to detect the ages (by considering intensity)

```
# Create sobel kernel for 2D images 

# First for X axis of image
sobel_x = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])

# Secondly for Y axis of image
sobel_y = np.array([[ -1, -2, -1], [ 0, 0, 0], [ 1, 2, 1]])

filtered_image = cv2.filter2D(gray,-1, sobel_y)

plt.imshow(filtered_image, cmap= 'gray')
```

```
# with 5*5 matrix 
sobel_z = np.array([[ -1,-1, -2,-1, -1], [0,0, 0, 0, 0], [ 1,1, 2, 1,1]])
filtered_image1 = cv2.filter2D(gray1,-1, sobel_z)
plt.imshow(filtered_image1, cmap= 'gray')
```

#My Observations:
## [1] Sobel (+) / (-) effect on image
#### Here, I found two observation that if total sum of sobel operator 0 then it works quite good. 
#### *But, if sum (+) results in increase whiteness (brighness = less detection of edge and mix with white background)
#### * On other hand, if (-) then , darkening of image results in less detaction and mixing with dark background


## [2] Effect of Sobel Operator Dimension
### Second Observation is relate to intensity of sobel operator, 
#### * If you use (3x3) sobel then change of sobel element affect image more than (10x10) operator. So, in your program if you want slite change to improve your machine vision then better to use higher dimension sobel operator.
